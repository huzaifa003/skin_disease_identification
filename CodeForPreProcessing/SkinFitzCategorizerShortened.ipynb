```python
#next step is to categorize all the images in each of the seperate folder so we can work individually on dataset

'''
So what we want here is to group each image by its labels so that we can use it as training set
let's start by identifying the labels, grouping them, getting their imageIds and then parse them in individual folders
The folder is in datasets and will be aned as categorized dataset. /content/drive/MyDrive/Datasets/FitzCategorized


Let's iterate the list and then finally group our images
'''

  labels = fitz17['label'].unique()
labels
```

```python
dataframes = {}
for label in labels:
  dataframes[label] = fitz17[fitz17['label'] == label]
  print(f"Total Images Approx for label {label} = {len(dataframes[label])}")

```

```python
sortedList = sorted(dataframes.items(), key= lambda x : len(x[1]), reverse=True) #sort the list based on df lengths for each understanding
for s in sortedList:
  print(f"The approx no of images for {s[0]} = {len(s[1])}")

# for label in labels:
#   print(f"Total Images Approx for label {label} = {len(sortedDicts[label])}")
```

```python
unexistingImages = []
oldFolderPath = "/content/drive/MyDrive/Datasets/FitsImages17/data/finalfitz17k"
folderPath = "/content/drive/MyDrive/Datasets/FitzCategorized"
for label in labels:
  dataframe = dataframes[label]
  titleColumn = dataframe['md5hash'].to_list()
  newFolderPath = os.path.join(folderPath, label)
  os.mkdir(newFolderPath) #Create a directory based on label


  for title in titleColumn:
    nam = title + ".jpg"
    picturePath = os.path.join(oldFolderPath, nam)
    if(os.path.exists(picturePath)):
      newPicturePath =  os.path.join(newFolderPath, nam)
      shutil.copy(picturePath, newPicturePath)
    else:
      unexistingImages.append({"name":nam, "label": label})

print(unexistingImages)
```
